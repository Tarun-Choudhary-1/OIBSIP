```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Unemployment in India.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df[' Date'] = pd.to_datetime(df[' Date'],format='mixed')
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=' Date', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

*Estimated Unemployment Rate (%) was increased between April 2020 to August 2020 because of first wave of Covid-19.*

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, x='Region', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by State')
plt.xticks(rotation=90)
plt.show()
```

*This boxplot shows the complete variation in Estimated Unemployment Rate(%) for each state and their median Estimated Uneployment Rate(%) according to data.*

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x=' Estimated Employed', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate vs. Employed Percentage')
plt.xlabel('Estimated Employed (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

*As the Estimated Employed(%) increases Estimated Unemployment Rate(%) decreases i.e. there is negative correlation between them.* 

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x=' Estimated Labour Participation Rate (%)', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate vs Labour Participation Rate')
plt.xlabel('Estimated Labour Participation Rate (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

*As the Estimated Labour Participation Rate(%) increases Estimated Unemployment Rate(%) increases i.e. there is positive correlation between them.*

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=' Date', y=' Estimated Unemployment Rate (%)', hue='Area')
plt.title('Unemployment Rate Over Time: Rural vs. Urban')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

*Urban area has more Estimated Unemployment Rate(%) than Rural area and both see a sudden spike during Covid-19 pandamic.*

```python
corr_matrix = df[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

*Positive correlation between Estimated Unemployment Rate (%) and Estimated Labour Participation Rate (%).*
*Negative correlation between Estimated Unemployment Rate (%) and Estimated Employed.*
